```python
#on importe la fonction Markdown
from IPython.display import Markdown

#une fonction qui va donner un nombre en binaire sans le 0b devant; ne marche pas, méthode compliquée
def bin8(nb):
    a=str(bin(nb))
    b= [0]*(len(str(bin(nb)))-2)
    for k in range(len(str(bin(nb)))-2):
        b[k]=a[k+2]
    return b

#la première ligne du tableau
tableau = "| dec \ bin | 128 | 64  | 32  | 16  |  8  |  4  |  2  |  1  |" + "\n"
tableau += "|:-:" * 9 + "|" + "\n"

#les lignes suivantes grâce à la fonction bin8()
for i in range(256) :
    binaire=bin8(i)
    tableau += f"| {i} |" 
    for i in range(len(binaire)-2):
        tableau+= f" {binaire[i+2]} |"
    tableau += "\n"

#on utilise la fonction Markdown pour afficher le tableau
Markdown(tableau)    
```
